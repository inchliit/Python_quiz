```python
import json
from IPython.display import display
from ipywidgets import Text, Button, VBox, Output

# Questions and answers based on the modules provided
quiz_data = [
    {"question": "What is the correct syntax to output 'Hello, World!' in Python?", "answer": "print(\"Hello, World!\")"},
    {"question": "Which character is used to define a comment in Python?", "answer": "#"},
    {"question": "True or False: Python is a case-sensitive language.", "answer": "True"},
    {"question": "What keyword is used to define a function in Python?", "answer": "def"},
    {"question": "How do you create a list in Python?", "answer": "Using square brackets []"},
    {"question": "What is the output of len([1, 2, 3])?", "answer": "3"},
    {"question": "Which operator is used for exponentiation in Python?", "answer": "**"},
    {"question": "What does the 'elif' keyword represent in Python?", "answer": "else if"},
    {"question": "What function is used to get the type of a variable?", "answer": "type()"},
    {"question": "True or False: Tuples are mutable.", "answer": "False"},
    {"question": "What is the output of 3 * 'Hi'?", "answer": "HiHiHi"},
    {"question": "Which Python collection type is unordered and does not allow duplicates?", "answer": "Set"},
    {"question": "What does the continue statement do in a loop?", "answer": "Skips the current iteration and continues with the next"},
    {"question": "How do you access the value associated with a key in a dictionary?", "answer": "Using the key inside square brackets"},
    {"question": "What is the purpose of the 'return' statement in a function?", "answer": "To send back a value from the function"},
    {"question": "What will be the output of print(10 // 3)?", "answer": "3"},
    {"question": "Which module is used for mathematical functions like sqrt?", "answer": "math"},
    {"question": "True or False: Strings in Python are immutable.", "answer": "True"},
    {"question": "What does the split() method do for a string?", "answer": "Splits the string into a list"},
    {"question": "What is the syntax to create a dictionary in Python?", "answer": "Using curly braces {key: value}"}
]

# Display quiz questions and input fields
questions_widgets = []
answers = {}
output = Output()

for i, q in enumerate(quiz_data):
    question_text = f"Q{i + 1}: {q['question']}"
    text_box = Text(placeholder=f"Answer for Q{i + 1}")
    questions_widgets.append((question_text, text_box))
    answers[f"Q{i + 1}"] = q['answer']

# Submit button functionality
def submit_answers(_):
    student_answers = {}
    for idx, (q_text, widget) in enumerate(questions_widgets):
        student_answers[f"Q{idx + 1}"] = widget.value
    
    # Compare answers
    results = {}
    correct_count = 0
    for key, correct_answer in answers.items():
        student_answer = student_answers.get(key, "").strip()
        if student_answer.lower() == correct_answer.lower():
            results[key] = "Correct"
            correct_count += 1
        else:
            results[key] = f"Incorrect (Correct: {correct_answer})"

    # Display results
    with output:
        output.clear_output()
        print("Quiz Results:")
        for key, result in results.items():
            print(f"{key}: {result}")
        print(f"Total Score: {correct_count}/{len(quiz_data)}")

submit_button = Button(description="Submit Answers", button_style='success')
submit_button.on_click(submit_answers)

# Arrange in a vertical layout
quiz_layout = VBox([
    VBox([
        VBox([Text(value=q_text, disabled=True), widget]) for q_text, widget in questions_widgets
    ]),
    submit_button,
    output
])

# Display the quiz layout
display(quiz_layout)
```
